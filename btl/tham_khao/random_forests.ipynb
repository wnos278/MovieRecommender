```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
dataset = 'E:\\20201\\ai\\MLDM-20-09-docs\\Bai-tap\\1-training-data.csv'
data = pd.read_csv(dataset,encoding= 'utf-8',sep=',',
                       names=["0", "1", "2", "3", "4","5","6","7", "8"], na_values='?')
data = data[1:]
```

```python
data
```

```python
type(data)
```

```python
data.fillna(value=0, inplace=True)
```

```python
data
```

```python
x = data[["0", "1", "2", "3", "4", "5", "6", "7"]]
y = data["8"]
```

```python
y.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
clf.predict([[0.45653953,3.451349732,65,-1.459023404,0,-1.511113075,6,0]])
```
